# Introduccion a la ciencia de datos

```python
import pandas as pd
ruta_csv = 'https://raw.githubusercontent.com/gitRaulMartinez/pythondatas/main/Padron%20electoral%20-%20Dataset.csv'
df = pd.read_csv(ruta_csv)
```

## Analisis multivariado

```python
sns.set()
data = df.loc[(df['Hora'] == '19:00')]
sns.boxplot(x=data['N° del Establecimiento'], y=data['Cantidad de votantes'])
plt.title('Cantidad de votantes por establecimiento')
plt.show()
```

```python
data = df.loc[(df['Hora'] == '19:00')]
total = data['Cantidad de votantes'].sum()
count = data['Cantidad de votantes'].count()

etiquetas = ['Votaron','No votaron']
valores = [total,abs(count * 348 - total)]

# Crear el gráfico de torta utilizando Seaborn
plt.figure(figsize=(6, 6))
sns.set(style='whitegrid')
sns.color_palette("pastel")
plt.pie(valores, labels=etiquetas, autopct='%1.1f%%', startangle=90)

# Configuraciones adicionales
plt.axis('equal')  # Para que el gráfico de torta sea un círculo en lugar de una elipse
plt.title('Participacion general de todo los establecimientos')

# Mostrar el gráfico
plt.show()
```

```python
# Datos de ejemplo
establecimientos =  df['N° del Establecimiento'].drop_duplicates().tolist()
valores = []
total = 348
for establecimiento in establecimientos:
  votos = df.loc[(df.iloc[:, 0] == establecimiento) & (df['Hora'] == '19:00')]['Cantidad de votantes'].mean()
  porncentaje = (votos / total) * 100
  valores.append(porncentaje)
# Crear el diagrama de barras utilizando Seaborn
sns.barplot(x=establecimientos, y=valores)

# Configuraciones adicionales
plt.xlabel('N° de establecimiento')
plt.ylim(0, 100)
plt.ylabel('% de Participación')
plt.title('Porcentaje de participación por cada establecimiento')

# Mostrar el diagrama de barras
plt.show()
```

```python
etiquetas = ['Votos validos','Votos en blanco','Votos nulos','Sin votar']

data = df.loc[(df['Hora'] == '19:00')]
total = data.groupby('Nombre del Establecimiento')['Cantidad de votantes'].sum()

etiquetas = total.keys().tolist()
valores = total.values.tolist()

# Crear el gráfico de torta utilizando Seaborn
plt.figure(figsize=(6, 6))
sns.set(style='whitegrid')
sns.color_palette('bright')
plt.pie(valores, labels=etiquetas, autopct='%1.1f%%', startangle=90)

# Configuraciones adicionales
plt.axis('equal')  # Para que el gráfico de torta sea un círculo en lugar de una elipse
plt.title('Distribución de votos por establecimientos')

# Mostrar el gráfico
plt.show()
```

## Analisis de correlación 

```python
sns.set()
# matriz de correlación y mapa de calor
columnas_incluidas = df.columns.difference(['N° del Establecimiento', 'Mesa','Total de votantes','Cantidad de votantes','Votos nulos','GOB-Votos en blanco','SEN-Votos en blanco','DIP-Votos en blanco','INT-Votos en blanco','CNJ-Votos en blanco'])

df_renombrado = df[columnas_incluidas].rename(columns={'GOB-Votos afirmativos': 'Gobernador', 'SEN-Votos afirmativos': 'Senador', 'DIP-Votos afirmativos':'Diputados','INT-Votos afirmativos': 'Intendente', 'CNJ-Votos afirmativos':'Concejal'})

correlaciones = df_renombrado.corr()

ax = sns.heatmap(correlaciones, annot=True, fmt='.2f')
plt.title('Votos afirmativos')
plt.show()
```
